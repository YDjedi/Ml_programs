```python
import numpy as np
import pandas as pd
import seaborn as sns

import category_encoders as ce
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv('megaGymDataset.csv')
```

```python
df.drop(columns=['Unnamed: 0', 'Title', 'Desc', 'RatingDesc', 'Rating' ], inplace=True)
```

```python
df
```

```python
def classcount(dataframe):
    for i in dataframe.columns:
        print(f"========{i}=========")
        print(dataframe[i].value_counts())
```

```python
classcount(df)
```

```python
x = df.drop(axis=1, columns='Level')
```

```python
x
```

```python
y = df['Level']
```

```python
y
```

```python
ohe = ce.OneHotEncoder()
res = ohe.fit_transform(x)
```

```python
res
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(res, y, test_size=0.2)
```

```python
random_forr = RandomForestClassifier()
random_forr.fit(X_train, Y_train)
random_forr.score(X_test, Y_test)
```

```python
Y_pred = random_forr.predict(X_test)
type(Y_pred)
```

```python
Y_test_arr = np.array(Y_test)
```

```python
test_actual_pred = np.vstack((Y_test_arr, Y_pred))
```

```python
act_pred = pd.DataFrame(test_actual_pred.T, columns=['actual', 'predicted'])
```

```python
act_pred
```

```python
cm = confusion_matrix(Y_test_arr, Y_pred)
cm
```
